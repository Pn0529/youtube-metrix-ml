```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import pickle
import gradio as gr
```

```python
df = pd.read_csv('/content/youtube_metrics (1).csv')
if 's.no' in df.columns:
    df.drop(columns=['s.no'], inplace=True)
```

```python
df['likes_per_view'] = df['likes'] / (df['views'] + 1)
df['log_views'] = np.log1p(df['views'])
```

```python

target_col = 'log_views'
X = df.drop(columns=['views', 'log_views'])
y = df[target_col]

num_cols = X.select_dtypes(include=['float64', 'int64']).columns.tolist()
cat_cols = X.select_dtypes(include=['object']).columns.tolist()
```

Replace `your_library_name` with the actual name of the library you want to install. For example, to install `requests`, you would use `pip install requests`.

```python


preprocessor = ColumnTransformer(
    transformers=[
        ('num', MinMaxScaler(), num_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ]
)

rf = RandomForestRegressor(random_state=42)
param_grid = {
    'n_estimators': [200, 500],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}

grid = GridSearchCV(rf, param_grid, cv=5, scoring='r2', n_jobs=-1)
model = Pipeline(steps=[('preprocessor', preprocessor),
                        ('regressor', grid)])
model.fit(X, y)

best_rf = grid.best_estimator_
scores = cross_val_score(best_rf, preprocessor.fit_transform(X), y, cv=5, scoring='r2')
print("Average CV R2:", scores.mean())

with open('best_model.pkl', 'wb') as file:
    pickle.dump(model, file)

print("Best model saved as 'best_model.pkl'")

original_df = pd.read_csv('/content/youtube_metrics (1).csv')
```

```python
def predict_outcome(*args):
    input_data = pd.DataFrame([args], columns=cat_cols + num_cols)
    if 'likes' in input_data.columns and 'views' in input_data.columns:
        input_data['likes_per_view'] = input_data['likes'] / (input_data['views'] + 1)
    if 'comments' in input_data.columns and 'views' in input_data.columns:
        input_data['comments_per_view'] = input_data['comments'] / (input_data['views'] + 1)
    input_data = input_data[X.columns]
    prediction = model.predict(input_data)
    prediction = np.expm1(prediction)
    return f"Predicted Views: {prediction[0]:.0f}"

ui_inputs = []
for col in cat_cols:
    unique_vals = original_df[col].dropna().unique().tolist()
    ui_inputs.append(gr.Dropdown(choices=unique_vals, label=f"{col}"))
for col in num_cols:
    ui_inputs.append(gr.Number(label=f"{col}"))

iface = gr.Interface(
    fn=predict_outcome,
    inputs=ui_inputs,
    outputs="text",
    title="YouTube Views Prediction App",
    description="Enter feature values to predict expected views."
)

iface.launch()
```
